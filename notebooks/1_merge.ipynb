```python
import pandas as pd
import glob
from os import listdir
```

```python
datasets_loc = glob.glob( '../datasets/*/**.csv' )
```

```python
datasets_loc
```

```python
df = None
for loc in datasets_loc:
    if df is None:
        df = pd.read_csv(loc)
        df['emoji'] = loc.split("/")[2]
    else:
        df_temp = pd.read_csv(loc)
        df_temp['emoji'] = loc.split("/")[2]
        df = df.append(df_temp)
```
